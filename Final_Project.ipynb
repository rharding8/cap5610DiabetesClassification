```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import optimizers, activations
from ucimlrepo import fetch_ucirepo
from imblearn.under_sampling import RandomUnderSampler
```

```python
data = pd.read_csv(r"C:\Users\migue\OneDrive\Documentos\UCF\Machine Learning\FinalProject\Source\Data\archive\diabetes_012_health_indicators_BRFSS2015.csv")
  
# data (as pandas dataframes) 
X = data.drop(columns=['Diabetes_012'])
y = data['Diabetes_012'] 
us = RandomUnderSampler(random_state=0)

x_resampled, y_resampled = us.fit_resample(X,y)

```

LSTM models work better with sequential data. Because of this, we will need to transform our data set a little to have some sort of "order" or sequence

```python
#scaling features
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x_resampled)

#encoding the target variable
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y_resampled)
y_cat = to_categorical(y_encoded)

# Reshape input to be 3D [samples, timesteps, features] for LSTM
# For simplicity, assume each record is a single timestep
X_reshaped = np.reshape(x_scaled, (x_scaled.shape[0], 1, x_scaled.shape[1]))



```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_cat, test_size=0.2, random_state=42)
```

```python
# number of neurons to test
n_neurons_test = [21,42,84,168]
learning_rates = [1e-4, 1e-3, 2.5e-3, 5e-3, 7.5e-3, 1e-2]


index = ['0.0001', '0.001', '0.0025', '0.005', '0.0075', '0.01']
neurons_21 = []
neurons_42 = []
neurons_84 = []
neurons_168 = []




for nn in n_neurons_test:
    for lr in learning_rates:    
        model = Sequential()
        model.add(LSTM(nn, input_shape=(X_reshaped.shape[1], X_reshaped.shape[2])))
        model.add(Dropout(0.2))
        model.add(Dense(3, activation= activations.sigmoid))

        model.compile(optimizer = optimizers.SGD(learning_rate= 0.001) , loss='categorical_crossentropy', metrics=['accuracy'])

        history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

        y_pred_prob = model.predict(X_test)
        y_pred = np.argmax(y_pred_prob, axis = 1)
        y_true = np.argmax(y_test, axis =1)

        precision  = precision_score (y_true, y_pred, average='weighted')
        recall = recall_score(y_true, y_pred, average='weighted')
        f1 = f1_score(y_true , y_pred, average='weighted')

        l1 = "Number of neurons: " + str(nn) + " & Learning Rate: " + str(lr)
        l2 = "Precision: " + str(precision)
        l3 = "Recall: " + str(recall)
        l4 = "F1 Score: " + str(f1)
        l5 = "==========================================================================================="
        
        if nn == 21:
            neurons_21.append(f1)
        elif nn == 42:
            neurons_42.append(f1)
        elif nn == 84:
            neurons_84.append(f1)
        elif nn ==168:
            neurons_168.append(f1)

        
        write = [l1,l2,l3,l4,l5]
        with open("Output.txt", "a") as f:
            for line in write:
                f.write(line + "\n")


        print(f'Number of Neurons: {nn:.2f} & learning rate: {lr:.4f}' )
        print(f'Precision: {precision:.2f}')
        print(f'Recall: {recall:.2f}')
        print(f'F1 Score: {f1:.2f}')
        print("==========================================================================================")


        path = "matrix\image_" + str(nn) + "_" + str(int(lr*10000))+".png"

        result = confusion_matrix(y_true , y_pred, normalize='pred')
        #plt.matshow(result,cmap = 'binary')
        disp = ConfusionMatrixDisplay(confusion_matrix=result)
        
        disp.plot(cmap='Blues')

        plt.title(l1)
        plt.savefig(path)
        plt.show()
        
        

df = pd.DataFrame({'21': neurons_21, '42': neurons_42, '84': neurons_84, '168':neurons_168}, index=index)
ax = df.plot.bar(rot=0)
```
